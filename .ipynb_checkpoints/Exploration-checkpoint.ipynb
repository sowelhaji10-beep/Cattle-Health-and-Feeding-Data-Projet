```python
# Vérifions  l’alignement des lignes 
assert len(base1) == len(base2)
assert (base1["Cattle_ID"] == base2["Cattle_ID"]).all()
assert (base1["Date"] == base2["Date"]).all()
assert (base1["Farm_ID"] == base2["Farm_ID"]).all()
```

```python
# fusion des deux dataset et ajout de la colonne Feeding_Frequency
df = base1.copy()
df["Feeding_Frequency"] = base2["Feeding_Frequency"]

df.shape
df.head()
```

```python
# Corrélation numérique globale
corr = df.select_dtypes(include=["int64", "float64"]).corr()

# Corrélations avec la cible lait
corr_target = corr["Milk_Yield_L"].sort_values(ascending=False)
corr_target.head(15)
```

```python
vars_corr = [
    "Milk_Yield_L", "Previous_Week_Avg_Yield",
    "Feed_Quantity_kg", "Feeding_Frequency",
    "Water_Intake_L", "Body_Temperature_C",
    "Heart_Rate_bpm", "Respiratory_Rate",
    "Body_Condition_Score", "Days_in_Milk"
]

plt.figure(figsize=(10, 8))
sns.heatmap(df[vars_corr].corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Corrélations autour du rendement laitier")

# Sauvegarde 
plt.savefig("correlations_rendement_laitier.png", dpi=300, bbox_inches="tight")
plt.show()
plt.close()

```
